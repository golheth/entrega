Las mediana Media y Moda de Los Precios

```python
# Calcular la media de los precios
media_precio = data1["price"].mean()

# Calcular la mediana de los precios
mediana_precio = data1["price"].median()

# Calcular la moda de los precios
moda_precio = data1["price"].mode().values[0]  # Usamos .values[0] para obtener el valor de la moda

print(f"Media de precios: {media_precio}")
print(f"Mediana de precios: {mediana_precio}")
print(f"Moda de precios: {moda_precio}")

```

Cuanto Facturaron las 10 Desarrolladoras Mas Grandes 

```python
# Agrupa los datos por desarrolladora y suma las ventas de cada una
ventas_por_desarrolladora = data1.groupby("publisher")["price"].sum()

# Ordena las desarrolladoras por ventas en orden descendente y selecciona las 10 más grandes
top_10_desarrolladoras = ventas_por_desarrolladora.nlargest(10)

# Configura el tamaño de la figura
plt.figure(figsize=(12, 6))

# Crea un gráfico de barras horizontal para las 10 desarrolladoras más grandes
sns.barplot(x=top_10_desarrolladoras.values, y=top_10_desarrolladoras.index, palette="viridis")

# Personaliza etiquetas y título
plt.xlabel("Ventas Totales")
plt.ylabel("Desarrolladora")
plt.title("Las 10 Desarrolladoras Más Grandes por Ventas Totales")

# Muestra el gráfico
plt.tight_layout()
plt.show()
```

```python
# Filtra los juegos con precio igual a 0
juegos_precio_cero = data1[data1['price'] == 0]

# Extrae el año de la columna "release_date" para estos juegos
juegos_precio_cero['release_year'] = juegos_precio_cero['release_date'].dt.year

# Elimina duplicados para contar juegos únicos por año
juegos_unicos_precio_cero_por_anio = juegos_precio_cero.drop_duplicates(subset=['release_year', 'game'])

# Cuenta la cantidad de juegos únicos con precio igual a 0 por año
cantidad_juegos_unicos_precio_cero_por_anio = juegos_unicos_precio_cero_por_anio['release_year'].value_counts().reset_index()
cantidad_juegos_unicos_precio_cero_por_anio.columns = ['Año', 'Cantidad de Juegos Únicos con Precio 0']

# Configura el tamaño de la figura
plt.figure(figsize=(12, 6))

# Crea un gráfico de líneas
plt.plot(cantidad_juegos_unicos_precio_cero_por_anio['Año'], cantidad_juegos_unicos_precio_cero_por_anio['Cantidad de Juegos Únicos con Precio 0'], marker='o', linestyle='-')

# Personaliza etiquetas y título
plt.xlabel("Año de Lanzamiento")
plt.ylabel("Cantidad de Juegos Únicos con Precio 0")
plt.title("Cantidad de Juegos Únicos con Precio 0 por Año")

# Muestra el gráfico
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
# Filtra los 10 principales publishers por facturación
top_10_publishers = data1['publisher'].value_counts().nlargest(10).index.tolist()

# Filtra el DataFrame solo para los juegos de los 10 principales publishers
data_top_publishers = data1[data1['publisher'].isin(top_10_publishers)]

# Calcula la facturación total de los juegos de los 10 principales publishers
facturacion_total_top_publishers = data_top_publishers.groupby('publisher')['price'].sum()

# Configura el tamaño de la figura
plt.figure(figsize=(8, 8))

# Crea un gráfico de torta
plt.pie(facturacion_total_top_publishers, labels=facturacion_total_top_publishers.index, autopct='%1.1f%%', startangle=140)

# Personaliza el título
plt.title("Proporción de Facturación Total por los 10 Principales Publishers")

# Muestra el gráfico
plt.tight_layout()
plt.axis('equal')  # Para asegurarse de que el gráfico de torta sea circular

plt.show()
```

En esta graficca se observa cuanto del total de facturacion se llevan estas publicadoras, cabe aclarar que no esta especificado en ningun año por lo que puede que algunas de ellas no existiesen en el pasado


```python
# Filtra los 10 principales publishers por facturación
top_10_publishers = data1['publisher'].value_counts().nlargest(10).index.tolist()

# Calcula la facturación total de los juegos de los 10 principales publishers
facturacion_total_top_publishers = data1[data1['publisher'].isin(top_10_publishers)]['price'].sum()

# Calcula la facturación total del resto de las empresas
facturacion_total_resto = data1[~data1['publisher'].isin(top_10_publishers)]['price'].sum()

# Configura el tamaño de la figura
plt.figure(figsize=(8, 8))

# Crea un gráfico de torta para mostrar la proporción
plt.pie([facturacion_total_top_publishers, facturacion_total_resto],
        labels=['Top 10 Publishers', 'Resto de Empresas'],
        autopct='%1.1f%%',
        startangle=140)

# Personaliza el título
plt.title("Proporción de Facturación Total entre Top 10 Publishers y Resto de Empresas")

# Muestra el gráfico
plt.tight_layout()
plt.axis('equal')  # Para asegurarse de que el gráfico de torta sea circular

plt.show()
```
